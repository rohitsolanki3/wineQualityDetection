```python
# Categorizing wine quality scores into 3 groups:
# 0 - Low (3 to 5), 1 - Medium (6), 2 - High (7 to 10)

def quality_group(score):
    if score <= 5:
        return 0  # Low
    elif score == 6:
        return 1  # Medium
    else:
        return 2  # High

df_wine_quality["quality_group"] = df_wine_quality["quality"].apply(quality_group)
```

```python
# Checking the distribution of wine quality categories (0: Low, 1: Medium, 2: High)
df_wine_quality["quality_group"].value_counts().sort_index()
```

```python
# Plotting class distribution with explicit hue to avoid deprecation warning
sns.countplot(x="quality_group", hue="quality_group", data=df_wine_quality, palette="Pastel1", legend=False)
plt.title("Class Distribution of Wine Quality Groups")
plt.xlabel("Quality Group")
plt.ylabel("Count")
plt.show()
```

```python
# Separating features and target variable (quality_group) for modeling
X = df_wine_quality.drop(columns=["quality_group"])
y = df_wine_quality["quality_group"]
```

```python
# Hyperparameter tuning for Random Forest
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    n_jobs=-1
)
grid_search.fit(X_train, y_train)

print("\nBest Parameters Found for Random Forest:", grid_search.best_params_)
optimized_rf = grid_search.best_estimator_
optimized_predictions = optimized_rf.predict(X_test)

print("Optimized Random Forest Accuracy:", round(accuracy_score(y_test, optimized_predictions) * 100, 2), "%")
```
